# Introduction to Principal Components Analysis

## Curse of Dimensionality

This happens when your data has a lot of features. A lot that its too much. This is a problem because of the following:

1. **Noise / Error**
    - With more features, we run the risk of introducing more noise to the dataset
2. **Scalability**
    - With more features, there is a lot more computations done. Because of this, training may take longer
3. **Difficulty in Analysis**
    - With more and more features, it gets more tedious and harder to actually analyze the data

## Principal Components Analysis

Principal Components Analysis (PCA) is a dimensionality reduction technique that tries to reduce the number of dimensions within the dataset by capturing patterns within the data in terms of its dimension. In short: It is some method to reduce the number of variables whilst keeping as much information as possible in terms of its variance.

<img src="pcaintro.png" width=500>

Source: https://environmentalcomputing.net/graphics/multivariate-vis/pca/


Let's go into the whiteboard to understand it better.

#### What can PCA do?

It learns some new set of variables to map the data points to which is able to keep as much information as possible from the dataset in terms of variance. These new axes are called Principal Components, and is the soul of PCA. These Principal Components, just like with X or Y or Z, is an axis that represents something new. Usually these Principal components are a combination of two axes and is meant to group features into something that makes it coherent. 

When capturing the "information" of the dataset, PCA captures information in the form of the variance. How much of this variance is actually captured per axis? That is what Principal Component does to actually learn these axes. Note that PCA captures axes and not lines. One main difference is that axes are orthogonal, meaning that there is no correlation between them. One cannot explain another. 

One thing that Principal Component Analysis does is translating datapoints into these new axes.

### Effects of Principal Components

#### Benefits

1. **Correlation**

Because Principal Components are Orthogonal to one another, there is 0 correlation to one another. This makes it perfect as features for a model that needs features that are independent (Naive Bayes, LinearRegression)

2. **New Dimensions and Summarized Features**

Principal Components are in themselves new representations of the points in the dimension space. It combines these features and maps the rows into these new dimensions. Effectively, we get a score for some new feature that may be more fit into predicting the value of the target.

3. **Less Noise**

Principal Components keep only the best features according to the captured variance. By reducing the number of features, we could've also removed features that just simply add noise into our dataset which makes it hard to get a generalized feature

-----

#### Cons

1. **May affect performance negatively**

Because we are removing some features, we are effectively removing their influence to the target. We are trading off performance for simplicity


### Assumptions of PCA

Source: https://www3.cs.stonybrook.edu/~mueller/teaching/cse564/Lec%2017%20-%20Principal%20Component%20Analysis.pdf

1. Linearity
    - Assumes that the dataset is linear
2. Importance of mean and covariance
    - High amount of variance captured != Discriminatory Features
3. Large Variances have important dynamics
    - Assumes components with larger variance correspond to interesting dynamics
    - Assumes components with lower variance correspond to noise
4. Continuous Data only
    - Since Principcal components look into capturing variance, PCA only properly works with Continuous data. Categorical Data is another story and requires a different method


# Fitting PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Get Features
X = df.drop(columns=['diagnosis'])

# Standardize Dataset
for col in X.columns:
    X[col] = StandardScaler().fit_transform(X[col].values.reshape(-1, 1))

# Create PCA
pca = PCA(n_components=X.shape[1])

# Fit Transform the Dataset
pca_X = pd.DataFrame(pca.fit_transform(X))

pca_X.head()
```

# Describing and Analyzing Principal Components

### Analysis 1: Understanding the influence of the features to the Principal Components

### Constructing the Component Matrix

```python
component_matrix = pd.DataFrame(pca.components_, columns=X.columns, index=[f'pca_{i}' for i in range(30)]).T
component_matrix
```

Scoring highly on a single component simply means that the original variables values for these features are explained by a single component.  **The higher the component score, the higher its influence over the PCA.**

With these names, let's rename our Principal Components

```python
# Get only first 5 components
pca_X = pca_X.iloc[:, :5]

# Rename the Columns
pca_X = pca_X.rename(columns={0: "Geometry", 1: "Fractal Dimension", 2: "Feel", 3:"Texture" , 4: "Perfectness"})

# Put Label
new_X = pca_X.copy()
new_X['diagnosis'] = df['diagnosis']

new_X
```

### Analysis 2: Understanding Explained Variance per Principal Component

------

#### The Explained Variance Ratio

As previously mentioned, the main goal of PCA is to be able to keep as much information as possible whilst creating new axes that represent the dataset very very very well. How is information set? In terms of its explained variance, which is how much variance of the entire dataset was captured by the principal component. If we have the same number of Principal Components and Columns in our original dataset, then we are supposed to get 1 by the end. Ideally, we would want to only keep a certain amount of features or else we destroy the purpose of PCA. 

```python
pd.DataFrame(pca.explained_variance_ratio_.cumsum())
```

```python
evr = pca.explained_variance_ratio_
cumsum_pca = pca.explained_variance_ratio_.cumsum()


plt.figure(figsize=(13, 4))
plt.plot([i for i in range(30)], cumsum_pca)
plt.grid(True, alpha=0.5)
plt.ylim([1, 0])
plt.show()
```
