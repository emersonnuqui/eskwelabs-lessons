## K-Means Clustering

### Initializing the dataset

To implement K-Means clustering, let's simulate a dataset.

```python
 # Generate fake dataset using scikit-learn
 features, clusters = make_blobs(
     n_samples = 1000,
     n_features = 10, 
     centers = 4,
     cluster_std = 0.4,
     shuffle = True
     )
```

By now, you should have a preliminary idea that we have 4 cluster centers. Do you think this will manifest in our elbow method later on?

```python
# Show features
features
```

```python
# Show clusters
clusters
```

```python
# Convert numpy array to pandas dataframe
cluster_data = pd.DataFrame(
    features,
    columns=[f"feature {i}" for i in range(1,11)]
    ) 
```

```python
# Show data
cluster_data
```

### Preprocessing the dataset

Before we proceed to clustering, we will check the following:
1. Does the data contain outliers?
2. Is the data standardized?

```python
# Create a box plot per feature
for column in cluster_data.columns:
  sns.boxplot(x=column, data=cluster_data)
  plt.title(column)
  plt.show()
```

Have you noticed an outlier?

While it is enticing to use an outlier detection (using Z-score or using interquartile range method), we must always treat an outlier for the WHOLE features with respect to the other data points.

We must also be VERY SURE about why a data point can be treated as an outlier since removing outliers can potentially degrade the quality of your data. Here's a story from NASA: https://robjhyndman.com/hyndsight/omitting-outliers/.




For now, let's skip removal of outliers and head straight to standardization.

```python
# Show data before scaling
cluster_data
```

```python
# Call scaler and apply to the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(cluster_data)

# Convert data to pandas
scaled_data = pd.DataFrame(scaled_data, columns=cluster_data.columns)
```

```python
# Show data after scaling
scaled_data
```

What happened?

We can see that each column is approximately of the same scale.

During scaling: for each column, we compute the mean and the standard deviation.

After, we convert each data point using the following transformation: z = (x - u) / s where z is the transformed value, x is the unscaled value, u is the mean for the specific column where the data belongs and s is the standard deviation.

### Applying K-Means clustering

Now, let's apply K-Means!

```python
# Initialize clustering
kmeans = KMeans(
    init="random",
    n_clusters=3,
    n_init=10,
    max_iter=300,
    random_state=42
    )
```

```python
# Apply clustering to the data
kmeans.fit(scaled_data)
```

```python
# Display the inertia (the lower, the better)
kmeans.inertia_
```

```python
# Display location of centroids
kmeans.cluster_centers_
```

To test different cluster numbers, let's create a function for clustering for modularity.

```python
# Create clustering function

def run_kmeans(n_clusters=None, scaled_data=None):
  # Initialize clustering
  kmeans = KMeans(
      init="random",
      n_clusters=n_clusters,
      n_init=10,
      max_iter=300,
      random_state=42
      )
  
  # Apply clustering to the data
  kmeans.fit(scaled_data)

  # Print intertia
  print(f"Inertia for n_clusters = {n_clusters}: {kmeans.inertia_}")

  return kmeans.inertia_, kmeans.labels_
```

```python
# Test clustering for cluster = 1
inertia, labels = run_kmeans(
    n_clusters=1,
    scaled_data=scaled_data
)
```

```python
# Test clustering for cluster = 2
inertia, labels = run_kmeans(
    n_clusters=2,
    scaled_data=scaled_data
)
```

```python
# Test clustering for cluster = 3
inertia, labels = run_kmeans(
    n_clusters=3,
    scaled_data=scaled_data
)
```

```python
# Test clustering for cluster = 4
inertia, labels = run_kmeans(
    n_clusters=4,
    scaled_data=scaled_data
)
```

```python
# Test clustering for cluster = 5
inertia, labels = run_kmeans(
    n_clusters=5,
    scaled_data=scaled_data
)
```

Lower inertia implies better clustering. What should be our number of clusters? Let's identify the optimal number of clusters using the elbow method!

Let's try number of clusters ranging from 1 to 10!

```python
# Create a container for inertia values
inertia_values = []
```

```python
# Run clustering and save inertia to our empty container
for k in range(1, 11):
  # Run our function
  inertia, labels = run_kmeans(
    n_clusters=k,
    scaled_data=scaled_data
  )

  # Save intertia to our container
  inertia_values.append(inertia)
```

With the help of matplotlib, let's visualize the data!

```python
# Set figure size
plt.figure(figsize=(6,3)  , dpi=200)

# Plot inertia vs number of clusters
plt.plot(range(1, 11), inertia_values)

# Format plot
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")

# Show plot
plt.show()
```

It seems that elbow occurs at 4 which checks out! Now, what's next? 

The most important part of clustering is in profiling: what makes one cluster different from the other? 

This is important since in a lot of decisions like in marketing, strategies will be informed by the profile of our clusters. 

```python
# Test clustering for cluster = 4
inertia, labels = run_kmeans(
    n_clusters=4,
    scaled_data=scaled_data
)
```

What's the labels?

This is the cluster of our data points ranging from 0 to n_clusters-1.

Note that in itself, these cluster labels will not make sense. We will be the one to interpret these labels later on.

```python
# Show labels
labels
```

```python
# Append labels to original data and scaled data
cluster_data['labels'] = labels
scaled_data['labels'] = labels
```

```python
# Show data
cluster_data
```

```python
# Show scaled data
scaled_data
```

### Profiling clusters

In profiling, we usually get the average of each feature per cluster. This way, we can compare the profile of one cluster to the other based on the average of the features.

```python
# Apply groupings in the ORIGINAL dataset
cluster_data.groupby('labels').mean()
```

Based on this, we can already spot the differences across clusters 0, 1, 2 and 3 based on the average of each of the feature columns. 

Now, let's go back to the scaled data. We will use them to build radar charts.

```python
# Apply groupings in the SCALED dataset
scaled_data.groupby('labels').mean()
```

```python
# Apply groupings in the dataset then transpose and save
grouped_data = scaled_data.groupby('labels').mean().T
```

```python
# Show the data
grouped_data
```

```python
# Show the transformed scaled data
grouped_data
```

```python
# Set figure size
plt.figure(figsize=(6,3), dpi=200)

# Plot radar chart per cluster
for i in range(0,4):
  # Filter cluster
  cluster = grouped_data[i]

  # Create radar chart
  fig = px.line_polar(
      cluster, 
      r=cluster.values, 
      theta=cluster.index, 
      line_close=True,
      title=f"Cluster {i}"
      )
  fig.update_traces(fill='toself')

  # Show figure
  fig.show()
```

## K-Modes Clustering

### Initializing the dataset

```python
# Show the data
data
```

```python
# Filter Philippines
philippine_data = data[
    data['economy']=='Philippines'
]
```

```python
# Show Philippine data
philippine_data
```

### Preprocessing the dataset

Let's cluster the data based on 3 features: education, income group and debit card ownership

```python
# Filter data
cluster_data = philippine_data[
    [
        'educ',
        'inc_q',
        'fin2'
    ]
]
```

```python
# Show data
cluster_data
```

For ease of analysis, let us make this dataset more interpretable.

To do this, let's head to the documentation at https://microdata.worldbank.org/index.php/catalog/3324/data-dictionary/F2?file_name=micro_world.

```python
# Create dictionaries for mapping
educational_group_mapping = {
    1: 'completed primary or less',
    2: 'secondary',
    3: 'completed tertiary or more'
}

income_group_mapping = {
    1:'poorest 20%',
    2:'second 20%',
    3:'middle 20%',
    4:'fourth 20%',
    5:'richest 20%'
}

debit_card_ownership_mapping = {
    1: 'yes',
    2: 'no',
    4: 'no'
}
```

```python
# Replace values using the mapping
cluster_data = cluster_data.replace(
    {
        'educ': educational_group_mapping,
        'inc_q': income_group_mapping,
        'fin2': debit_card_ownership_mapping
     }
    )
```

```python
# Show data
cluster_data
```

```python
# Check nulls
cluster_data.isna().sum()
```

```python
# Rename columns
cluster_data = cluster_data.rename(
    columns = {
        'educ':'educational_background', 
        'inc_q':'income_group',
        'fin2': 'has debit card'
        }
)
```

```python
# Show data
cluster_data
```

### Applying K-Modes clustering

The module kmodes behaves similarly like K-Means. Let's create a function for running K-Modes.

```python
# Create clustering function
def run_kmodes(n_clusters=None, cluster_data=None):
  # Initialize clustering
  kmodes = KModes(
      init="random",
      n_clusters=n_clusters,
      random_state=42
      )
  
  # Apply clustering to the data
  kmodes.fit(cluster_data)

  # Print intertia
  print(f"Cost for n_clusters = {n_clusters}: {kmodes.cost_}")

  return kmodes.cost_, kmodes.labels_
```

Let us cluster the dataset using n_clusters ranging from 1 to 10.

Notice that in this case, we use cost instead of inertia. Clustering cost, in K-Modes, is defined as the sum distance of all points to their respective cluster centroids.

Idea: lesser cost, better clustering!

```python
# Create container for cost
cost_values = []
```

```python
# Run clustering and save inertia to our empty container
for k in range(1, 11):
  # Run our function
  cost, labels = run_kmodes(
    n_clusters=k,
    cluster_data=cluster_data
  )

  # Save cost to our container
  cost_values.append(cost)
```

Let's plot our dataset to find the "optimal" number of clusters.

```python
# Set figure size
plt.figure(figsize=(6,3)  , dpi=200)

# Plot inertia vs number of clusters
plt.plot(range(1, 11), cost_values)

# Format plot
plt.xticks(range(1, 11))
plt.xlabel("Number of Clusters")
plt.ylabel("Cost")

# Show plot
plt.show()
```

It is difficult to choose an optimal cluster because the elbow is not pronounced.

However, note that elbow method is only our guiding light. In the end, clustering is a somewhat subjective task.

For now, let's assume that the optimal cluster is 4.

### Profiling clusters

```python
# Run our K-Modes function for n_clusters=4
cost, labels = run_kmodes(
  n_clusters=4,
  cluster_data=cluster_data
)
```
